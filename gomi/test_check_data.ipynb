```python
import pandas as pd
```

```python
df = pd.read_csv('../datas/real_estate_data_20211_20224_14.csv')
```

```python
df.columns
```

```python
def pv_count(df_, target_col):
    df_pv_tmp = df_.pivot_table(index=target_col, values='Area', aggfunc='count').sort_values('Area', ascending=False)
    display(df_pv_tmp.T.iloc[:,:15])
```

```python
cols = ['Type', 'Region','Municipality', 'LandShape','Structure', 'Use', 'Purpose', 'CityPlanning', 'Period', 'FloorPlan', 'Renovation']
for col in cols:
    pv_count(df, col)
```
