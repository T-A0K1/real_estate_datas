```python
def pv_count(df_, target_col):
    df2_ = df_.loc[:,['Area', target_col]].fillna('Nan')
    df_pv_= df2_.pivot_table(
        index=target_col, 
        values='Area', 
        aggfunc='count',
        fill_value=0
        ).sort_values('Area', ascending=False)
    
    df_pv_['Percentage'] = (df_pv_['Area'] / df_pv_['Area'].sum() * 100).round(1)
    df_pv_.columns = [target_col+'_count', target_col+'(%)']
    display(df_pv_.T.iloc[:,:15])
```

```python
cols = ['Type', 'Region','Municipality', 'LandShape','Structure', 'Use', 'Purpose', 'CityPlanning', 'Period', 'FloorPlan', 'Renovation']
for col in cols:
    pv_count(df, col)
```
